#Problem - 4: Matrix Operations:
• For the following arrays:
A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
Prove following with Numpy:
1. Prove A.A−1 = I.
2. Prove AB ̸= BA.
3. Prove (AB)T = BTAT

```python
import numpy as np
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

Inverse_of_A=np.linalg.inv(A)
Identity=np.dot(A,Inverse_of_A)
print(Identity)
print("\n")
AB=np.dot(A,B)
BA=np.dot(B,A)
print("A * B =", AB)
print("B * A =", BA)
print("\n")
transpose_AB=np.transpose(AB)
transpose_A=np.transpose(A)
transpose_B=np.transpose(B)
A_transpose_B_transpose=np.dot(transpose_B,transpose_A)
print("Transpose of AB=",transpose_AB)
print("Transpose of A and B=",A_transpose_B_transpose)
```

#• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}

```python
import numpy as np
A=np.array([[2,-3,1],[1,-1,2],[3,1,-1]])
B=np.array([-1,-3,9])
A_Inverse=np.linalg.inv(A)
X=np.dot(A_Inverse,B)
print(X)
print("\n")
X2=np.linalg.solve(A,B)
print(X2)
```
